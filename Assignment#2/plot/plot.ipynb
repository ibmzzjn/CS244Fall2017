```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = np.genfromtxt("data.csv", dtype=int, delimiter=',', names=True) 
IR1 = data['IR1']
RED1 = data['RED1']
IR2 = data['IR2']
RED2 = data['RED2']
IR3 = data['IR3']
RED3 = data['RED3']
IR4 = data['IR4']
RED4 = data['RED4']



# Four axes, returned as a 2-d array
f, axarr = plt.subplots(1, 2)
axarr[0].plot(RED1)
axarr[0].set_title('RD1, current = 0.4mA')
axarr[1].plot(IR1)
axarr[1].set_title('IR1, current = 0.4mA')

# Four axes, returned as a 2-d array
f, axarr = plt.subplots(1, 2)
axarr[0].plot(RED2)
axarr[0].set_title('RD2, current = 6.4mA')
axarr[1].plot(IR2)
axarr[1].set_title('IR2, current = 6.4mA')

# Four axes, returned as a 2-d array
f, axarr = plt.subplots(1, 2)
axarr[0].plot(RED3)
axarr[0].set_title('RD3, current = 25.4mA')
axarr[1].plot(IR3)
axarr[1].set_title('IR3, current = 25.4mA')

# Four axes, returned as a 2-d array
f, axarr = plt.subplots(1, 2)
axarr[0].plot(RED4)
axarr[0].set_title('RD4, current = 50.0mA')
axarr[1].plot(IR4)
axarr[1].set_title('IR4, current = 50.0mA')


# axarr[0, 1].plot(RED2)
# axarr[0, 1].set_title('RED2')

# axarr[1, 0].plot(RED3)
# axarr[1, 0].set_title('RED3')

# axarr[1, 1].plot(RED4)
# axarr[1, 1].set_title('RED4')



# # Four axes, returned as a 2-d array
# f, axarr = plt.subplots(1, 2)
# axarr[0, 0].plot(IR1)
# axarr[0, 0].set_title('IR1')

# axarr[0, 1].plot(IR2)
# axarr[0, 1].set_title('IR2')

# axarr[1, 0].plot(IR3)
# axarr[1, 0].set_title('IR3')

# axarr[1, 1].plot(IR4)
# axarr[1, 1].set_title('IR4')
```

```python
var_red1 = np.var(RED1)
var_red2 = np.var(RED2)
var_red3 = np.var(RED3)
var_red4 = np.var(RED4)
var_red = [var_red1, var_red2, var_red3, var_red4]

var_ir1 = np.var(IR1)
var_ir2 = np.var(IR2)
var_ir3 = np.var(IR3)
var_ir4 = np.var(IR4)
var_ir = [var_ir1, var_ir2, var_ir3, var_ir4]

current = [0.4,6.4, 25.4, 50]


plt.plot(current, var_red/sum(var_red), color="blue", linewidth=2.5, linestyle="-", label="RED")
plt.plot(current, var_ir/sum(var_ir), color="red",  linewidth=2.5, linestyle="-", label="IR")
plt.title("normalized sensor data for RED and IR variations across different current settings");
plt.legend(loc='lower right')
```
