```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

right = np.genfromtxt("team9_assignment4_right.csv", dtype=float, delimiter=',', names=True) 
left = np.genfromtxt("team9_assignment4_left.csv", dtype=float, delimiter=',', names=True) 
up = np.genfromtxt("team9_assignment4_up.csv", dtype=float, delimiter=',', names=True) 
down = np.genfromtxt("team9_assignment4_down.csv", dtype=float, delimiter=',', names=True) 
circular = np.genfromtxt("team9_assignment4_circular.csv", dtype=float, delimiter=',', names=True) 

RX = right['X']
RY = right['Y']
RZ = right['Z']

LX = left['X']
LY = left['Y']
LZ = left['Z']

UX = up['X']
UY = up['Y']
UZ = up['Z']

DX = down['X']
DY = down['Y']
DZ = down['Z']

CX = circular['X']
CY = circular['Y']
CZ = circular['Z']
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(511)
ax1.plot(RX)
ax1.set_title("right")
ax1.axis([0, 500, -0.1, 0.3]);

ax2 = fig.add_subplot(512)
ax2.plot(LX)
ax2.set_title("left")
ax2.axis([0, 500, -0.1, 0.3]);


ax3 = fig.add_subplot(513)
ax3.plot(UX)
ax3.set_title("up")
ax3.axis([0, 500, -0.1, 0.3]);

ax4 = fig.add_subplot(514)
ax4.plot(DX)
ax4.set_title("down")
ax4.axis([0, 500, -0.1, 0.3]);


ax5 = fig.add_subplot(515)
ax5.plot(CX)
ax5.set_title("circular")
ax5.axis([0, 500, -0.1, 0.3]);
```
